<h3>4.1.1 수치형 데이터</h3>

<b>수치형 데이터(numerical data)</b>는 사칙 연산이 가능한 데이터. data type으로는 int, float, bool 등이 있다. (bool은 0과 1로 표현되기에 사칙 연산이 가능.)


<li>연속형 데이터(continuous data): 데이터가 연속적으로 이어진다. 예를 들어, 키, 몸무게, 온도 등이 있다.</li>
<br>
<li>이산형 데이터(discrete data): 데이터가 이산적으로 이어진다. 예를 들어, 사람의 수, 소의 수, 차의 수 등이 있다.</li>
<br>

*연속형은 실수형, 이산형은 정수형 데이터*

<h3>4.1.2 범주형 데이터</h3>

<b>범주형 데이터(categorical data)</b>는 범주를 나눌 수 있는 데이터이며 사칙 연산이 불가능한 데이터. data type으로는 object, category 등이 있다.(여기서 object와 category는 pandas라이브러리에서 사용됨. object는 string type과 동일하지만, category는 고유한 값을 가진 데이터로 효율 적으로 메모리 관리 할 수 있께 해줌)


<li>순서형 데이터(ordinal data)</b>순위를 매길 수 있는 데이터. 예를 들어, 학점(A, B, C, D) 등이 있다.</li>
<br>
<li>명목형 데이터(nominal data)</b>순위가 따로 존재하지 않는 데이터. 예를 들어, 성별(남, 여) 등이 있다. 또한 우편번호 역시 숫자로 되어있지만 사칙연산과 관련 없고 순위가 존재하지 않기 때문에 명목형으로 분류된다.</li>
<br>



```python
import sys
import pandas as pd

obj_data0 = pd.Series(["코끼리", "사자", "여우", "토끼"])
cat_data0 = pd.Series(["코끼리", "사자", "여우", "토끼"], dtype='category')
obj_data1 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"])
cat_data1 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"], dtype="category")
obj_data2 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"])
cat_data2 = pd.Series(["코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼", "코끼리", "사자", "여우", "토끼"],
                      dtype="category")

print(sys.getsizeof(obj_data0))  # object1 타입의 메모리 사용량
print(sys.getsizeof(cat_data0))  # category1 타입의 메모리 사용량
print(sys.getsizeof(obj_data1))  # object1 타입의 메모리 사용량
print(sys.getsizeof(cat_data1))  # category1 타입의 메모리 사용량
print(sys.getsizeof(obj_data2))  # object2 타입의 메모리 사용량
print(sys.getsizeof(cat_data2))  # category2 타입의 메모리 사용량
```

<h3>4.3.1 히스토그램</h3>

수치형 데이터의 구간별 빈도수를 나타내는 그래프. (historical의 histo-와 gram의 합성어. 역사적 데이터 분포를 데이터 형태로 시각적으로 표현)

```python
sns.histplot(data= titanic, x= 'age')
```

<li>bins는 구간을 정의</li>
</br>
<li>hue는 데이터의 추가적인 범주형을 지정. 색조로 이를 표현</li>
</br>
<li>multiple은 여러개의 변수를 누적해서 표현</li>

```python
sns.histplot(data=titanic, x= "age", bins=10)
```

```python
sns.histplot(data=titanic, x= "age", hue='alive')
# 회색구간은 alive no와 yes의 공통 부분
```

```python
sns.histplot(data=titanic, x= "age", hue='alive', multiple = 'stack')
```

<h3>4.3.2 커널밀도추정 함수 그래프(kdeplot)</h3>

데이터 분포를 나태내는 함수로 특정 구간 데이터가 존재할 확률을 나타냄


<img src="../ch4/img/kde.png" alt="kde">


```python
sns.kdeplot(data=titanic, x='age')
```

```python
sns.kdeplot(data=titanic, x='age',hue='alive', multiple='stack')
```

<h3>4.3.3 분포도(displot)</h3>

kaggle에서 자주 사용하는 분포도

파라미터 조정을 통해서 hisplot()과 kdeplot()이 제공하는 기본 그래프를 모두 그릴 수 있음


```python
sns.displot(data=titanic, x='age', kind='kde')
```

```python
sns.displot(data=titanic, x='age', kde=True)
```

<h3>4.3.4 러그플롯(rugplot)</h3>

러그플롯은 주변 분포(marginal distribution)을 나태나는 그래프. 

```python
sns.rugplot(data=titanic, x = 'age')
```

```python
sns.kdeplot(data=titanic, x= 'age')
sns.rugplot(data=titanic, x = 'age')
```

```python
sns.rugplot(data=titanic, x = 'age')
sns.kdeplot(data=titanic, x= 'age')
```

<h3>4.4.1 막대그래프(barplot)</h3>

범주형 데이터 값에 따라 수치형 데이터 값의 평균과 신뢰구간을 그려줌.

평균은 막대의 높이로, 신뢰구간은 오차 막대로 표현

<h3>4.4.2 포인트 플롯(pointplot)</h3>

```python
sns.pointplot(x= 'class', y='fare', data=titanic)
```

<h3>4.4.3 박스플롯(boxplot)</h3>

다섯가지 요약 수치를 통해서 정보를 나타냄. (Q1, Q2, Q3, 최솟값, 최댓값)

1. 제1 사분위수(Q1): 전체 데이터 중 하위 25%에 해당하는 값

2. 제2 사분위수(Q2): 중앙값

3. 제3 사분위수(Q3): 전체 데이터 중 상위 25%에 해당하는 값

4. 최댓값: Q3 + (1.5 * IQR) 이하의 가장 큰 값

5. 최솟값: Q1 - (1.5 * IQR) 이상의 가장 작은 값

    - 사분위 범위 수(IQR): Q3 - Q1

    - 이상치: 최대값보다 큰 수, 최솟값보다 작은 수

<img src='./img/boxplot.png' alt="boxplot" style = "width: 75%; height: auto;">

```python
sns.boxplot(x='class', y = 'age', data=titanic)
```

<h3>4.4.4 바이올린플롯</h3>

바이올린 플롯은 박스플롯과 커널밀도추정 함수 그래프를 합쳐 놓은 그래프.

박스플롯이 제공하는 데이터를 커널밀도추정 함수 그래프 형태로 나타낸 것.

```python
sns.violinplot(x='class', y= 'age', hue='sex', data=titanic, split=True)
```

<h3>4.4.5 카운트플롯(countplot)</h3>

barplot과 거의 동일한 모습을 보이지만, barplot은 데이터의 평균값을 제공하는 대신 카운트플롯은 데이터의 개수를 파악할 수 있다.

따라서 x와 y의 변수 중 하나만 제공해야 한다.

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='class', y= 'age', data=titanic)
```

<h3>4.4.6 파이 그래프(pie)</h3>

파이그래프는 데이터를 비율로 알아보기 좋은 그래프이다. 하지만 seaborn에서는 pie그래프를 제공하지 않아, matplotlib을 사용해야 한다.

```python
import matplotlib.pyplot as plt

x = [10,60,30]
labels = ['A', 'B', 'C']

plt.pie(x= x, labels = labels, autopct='%.1f%%')
```

<h3>4.5.1 히트맵(heatmap)</h3>

데이터 간 관계를 색상으로 표현한 그래프. 비교할 데이터가 많을 때 사용.

```python
flights = sns.load_dataset('flights')

flights.head()
```

```python
flights_pivot = flights.pivot(index='month',
                              columns='year',
                              values='passengers')
flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
```

<h3>4.5.2 라인플롯(lineplot)</h3>

두 변수간 관계의 변화를 보여줄때 사용.

x 파라미터에 전달한 값에 따라 y파라미터에 전달한 값의 평균과 95%의 신뢰구간(ci로 변경 가능)을 나타냄

```python
sns.lineplot(x='year', y='passengers', data=flights)
```

<h3>4.5.3 산점도(scatterplot)</h3>

두 데이터의 관계를 점으로 표현하는 그래프

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.scatterplot(x= 'total_bill', y= 'tip', data=tips)
```

```python
sns.scatterplot(x= 'total_bill', y= 'tip', hue='size' ,data=tips)
```

<h3>회귀선을 포함한 산점도 그래프(regplot)</h3>

regplot()은 산점도와 선형 회귀선을 동시에 그려주는 함수(ci는 신뢰구간을 의미)

```python
sns.regplot(x = 'total_bill', y= 'tip', data=tips, ci= 90)
```
